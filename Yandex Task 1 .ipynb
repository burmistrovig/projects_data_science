### 1.5. Movie_duration_category

```python
# For creatin better and simplier world,  let's make once extra series object for thi function
movie_category_series = []
def duration_category(dataframe):
    duration_cat = dataframe['duration']
    for i in range(len(dataframe)):
        if duration_cat[i] < 90:
            movie_category_series.append("1. <90")
        elif duration_cat[i] >90 and duration_cat[i] < 120:
            movie_category_series.append("2. 90-120")
        else:
            movie_category_series.append("3. >120")
```

```python
duration_category(df)
```

```python
df['movie_duration_category'] = movie_category_series
```

Checking results of our function.

```python
df['movie_duration_category'].value_counts()
```

### 1.8 Movie_plot_category

```python
love = df["plot_keywords"].str.contains(r"\blove\b")
death = df["plot_keywords"].str.contains(r"\bdeath\b")

df["movie_plot_category"] = \
    np.where(love & death,
             "love_and_death",
             np.where(love,
                      "love",
                      np.where(death,
                               "death", 
                               "other")))
```
